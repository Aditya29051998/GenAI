```python
import pandas as pd
#Initial data
Reviews = ["I love this product!",
           "It's amazing and exceeded my expectations.",
           "Highly recommended!",
           "This is the best purchase I've made in a while.",
           "Great value for the money.",
           "Not bad, could be better.",
           "I'm disappointed with this product.",
           "It didn't meet my expectations.",
           "I wouldn't recommend it.",
           "This is a terrible product."]
df_reviews = pd.DataFrame(Reviews, columns=['Reviews'])
print(df_reviews)
df_reviews['Reviews_Lowercase'] = df_reviews['Reviews'].str.lower()
df_reviews
```

Removing Punctuation and Emojis

```python
import re
df_reviews['Reviews_No_Punctuation'] = df_reviews['Reviews_Lowercase'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
df_reviews
```

Remove Stop Words


```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
df_reviews['Reviews_No_Stopwords'] = df_reviews['Reviews_No_Punctuation'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
df_reviews
```

Stemming

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
df_reviews['Reviews_Stemmed'] = df_reviews['Reviews_No_Stopwords'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split()]))
df_reviews
```

Tokenization

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
word_tokenize(df_reviews['Reviews_Stemmed'][0])
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
df_reviews['Reviews_Tokenized'] = df_reviews['Reviews_Stemmed'].apply(lambda x: word_tokenize(x))
df_reviews
```

Sentence TokenizerI

```python
from nltk.tokenize import sent_tokenize
df_reviews['Reviews_Subword_Tokenized'] = df_reviews['Reviews'].apply(lambda x: sent_tokenize(x))
df_reviews
```

```python
!pip install gensim
from gensim.models import Word2Vec

# Sample Text
sentences = [["King", "Queen", "man", "Women"],
            ["apple", "bananna", "fruit"],
            ["king", "man", "Ruler"],
            ["queen", "woman", "ruler"]]  # Sentences should be a list of lists

# Train Word2Vec model
model = Word2Vec(sentences, vector_size=3, window=2, min_count=1, sg=1)
word_vectors = model.wv
```

```python
print("Embedding for king:",model.wv['king'])
print("Most similar words to king:",model.wv.most_similar('king'))
print("Embedding for Queen:",model.wv['Queen'])
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

#Word to visualize
words = ["King", "Queen", "man", "Women","bananna"]
word_vectors = [model.wv[word] for word in words]

#PCA
pca = PCA(n_components=2)
word_vectors_2d = pca.fit_transform(word_vectors)

#Plot
plt.figure(figsize=(10, 10))
for i, word in enumerate(words):
    plt.scatter(word_vectors_2d[i, 0], word_vectors_2d[i, 1])
    plt.annotate(word, (word_vectors_2d[i, 0], word_vectors_2d[i, 1]))
    plt.text(word_vectors_2d[i, 0], word_vectors_2d[i, 1], word)
plt.title("Word Embeddings Visualization")
plt.show()
```
